```python
data["Embarked"] = data["Embarked"].fillna('S')# filling the missing values of Embarked with mode
```

## 5. Check and handle outliers in at least 3 columns in the datase

######  1.checking 'Age' column

```python
plt.boxplot(data['Age'])# checking 'Age' column
plt.title('Box plot  of Age  (before handling outliers)')
```

```python
Q1=np.quantile(data['Age'],0.25)
Q3=np.quantile(data['Age'],0.75)
print('Q1,Q3=')
print(Q1,'\n',Q3)
IQR=Q3-Q1
print('IQR=',IQR)

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print('lower_limit=',lower_limit)
print('upper_limit=',upper_limit)

# finding outliers
outlier=[]
for a in data['Age']:
    if (a<lower_limit or a>upper_limit):
        outlier.append(a)
print('\noutlier=',outlier)

```

* There are outliers in Age below lower limit and above upper limit.

```python
# finding index of outliers 
index_of_outliers=data.loc[data['Age'].gt(upper_limit) | data['Age'].lt(lower_limit)].index
print('index_of_outliers\n')
print(index_of_outliers)
```

```python
#removing outliers
data.drop([8,  12,  16,  17,  34,  55,  79,  95,  97, 117, 120, 153, 165,
            171, 173, 175, 184, 196, 206, 233, 253, 269, 276, 281, 298, 306,
            327, 341, 367, 382, 387, 439, 457, 468, 470, 480, 484, 488, 493,
            494, 531, 546, 556, 571, 588, 626, 627, 631, 643, 645, 648, 660,
            673, 685, 695, 746, 756, 773, 789, 804, 825, 828, 830, 832, 852,
            880],inplace=True)
#displaying the data
data
```
